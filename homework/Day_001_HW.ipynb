1. 核心問題為何<br>
若我需要提升車隊業績，我將試著尋找乘客量最多的路線以及時段並合理安排車隊分布，因此首要分析為找出最佳路線及最佳時間。<br><br>
2. 資料從何而來<br>
以問卷調查方式在不同地區、不同時段的路人作為樣本，問卷中需提及搭車時段及最常上車地區，且需分散在某一區不同地點避免資料產生誤差，再進一不向車隊司機蒐集不同區域及早晚之實際載客人數。<br><br>
3. 蒐集而來的資料型態為何<br>
問卷資料應為結構式資料，搭車時段以早、晚作為選擇，上車地區以區域作為選擇，各項數據次數做加總得到數值資料，並結合司機所得實際資料作為dataset。<br><br>
4. 這組資料想解決的問題如何評估<br>
可利用此dataset製作二種confusion matrix，一為區域的confusion matrix，另一為時段的confusion matrix，並計算Recall或MCC來判斷和時段或區域更適合派出大量車隊。

```python
#作業3
import numpy as np#矩陣函數
import matplotlib.pyplot as plt #資料視覺化
```

```python
def mean_squared_error(y,yp):
    mse = MSE = (sum(y-yp)**2)/len(y)
    return mse
```

```python
def mean_absolute_error(y, yp):
    mae = MAE = sum(abs(y - yp)) / len(y)#避免後面取值衝突故用小寫來return
    return mae
```

```python
w = 3
b = 0.5

x_lin = np.linspace(0, 100, 101)#創建等差數列(0-100共101個)

y = (x_lin + np.random.randn(101) * 5) * w + b #random.randn生成0-1間的亂數,一維共101個#生成實際值

plt.plot(x_lin, y, 'b.', label = 'data points')#畫圖
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()
```

```python
y_hat = x_lin * w + b#生成預測值
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
# 執行 Function, 確認有沒有正常執行
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))#.3f為取到第3個小數點
print("The Mean absolute error is %.3f" % (MAE)) 
```
